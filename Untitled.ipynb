```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import plotly.express as px
import datetime
```

```python
df = pd.read_csv('Significant_Earthquakes.csv')
df.tail()
```

```python
df.drop('Unnamed: 0',axis=1 , inplace=True)
```

```python
df.info()
```

```python
df['time']=pd.to_datetime(df['time'])
```

```python
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
df['weekday'] = df['time'].dt.weekday
df.head()
```

```python
df['depth'] = df['depth'].fillna(df['depth'].mean())
```

```python
px.histogram(data_frame=df,x='year',color='type')
```

```python
fig=px.bar(x=df.groupby('type')['mag'].mean().index,y=df.groupby('type')['mag'].mean().values,color=df.groupby('type')['mag'].mean().index,title='Average Mag for Each Type')
fig.update_layout(
    xaxis_title='Type',
    yaxis_title='mag scale'
)
fig.show()
```

```python
fig=px.bar(x=df.groupby('type')['mag'].max().index,y=df.groupby('type')['mag'].max().values,color=df.groupby('type')['mag'].max().index,title='Max Mag for Each Type')
fig.update_layout(
    xaxis_title='Type',
    yaxis_title='mag scale'
)
fig.show()
```

```python
df[df['mag']==df['mag'].max()]
```

```python
fig=px.bar(x=df.groupby('type')['depth'].mean().index,y=df.groupby('type')['depth'].mean().values,color=df.groupby('type')['depth'].mean().index,title='mean depth for Each Type')
fig.update_layout(
    xaxis_title='Type',
    yaxis_title='depth'
)
fig.show()
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation='vertical')
sns.barplot(df['net'].value_counts().index,df['net'].value_counts().values)
plt.show()
```

```python
plt.figure(figsize=(20,10),dpi=100)
plt.xticks(rotation='vertical')
sns.barplot(df['locationSource'].value_counts().index,df['locationSource'].value_counts().values)
plt.show()
```

```python
sns.scatterplot(data=df,x='mag',y='depth')
plt.show()
```
